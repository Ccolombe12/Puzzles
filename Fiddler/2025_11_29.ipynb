How many ordered triples have the property that they make every subset of $T$?

Note that we need at least $N_T = \lfloor\log_2(T) + 1\rfloor$ bits to describe all the numbers $[T]$


```python
from itertools import combinations
import numpy as np
from math import floor
from matplotlib import pyplot as plt


# need a function that checks if we can make all the numbers
def count_min_set(T):
    N = floor(np.log2(T) + 1)

    def check(candidate_nums, T):
        """Return True if the set of numbers can make {1,...,T}"""
        can_make = 0
        for v in candidate_nums:
            if v > can_make + 1:
                return False
            can_make += v
        return can_make >= T

    vals = list(range(3, T))  # the set of numbers we can pull from
    total = 0
    for rest in combinations(vals, N - 2):
        candidate = [1, 2] + list(rest)  # ensure we have 1 and 2.
        total += check(candidate, T)

    return total


x = list(range(3, 101))
y = [count_min_set(v) for v in x]
print(f"There are {y[7]} tuples that make T = {10}")
print(f"there are {y[-1]} tuples that make T = {100}")
plt.plot(x, y)
plt.yscale("log")
plt.show()
```
