# Part 1


This problem is all about equilateral triangles on a sphere (since the legs all have length $s$, we walk in three straight lines, and need to return to the starting point). Some quick searching yields this excellent article at https://www.usna.edu/Users/physics/mungan/_files/documents/Scholarship/SphericalTriangle.pdf. 

Which proposes we look at the problem using this diagram:


Where the bottom leg is along the equator of the circle ($z = 0$), and the first leg makes an $\alpha$ angle from the horizon. 

Here we must imagine $\alpha = \pi / 3$ and that the starting point on the red line to the left of the $x = 0$ axis line.

If we normalize this to be the unit sphere, the distance traveled along a great-circle $s$ will take up exactly $s$ radian. (for example the path from the starting point along the red line is $s$ units on the surface and took $s$ radian to complete.)

The canted great circle in black is the circle in which our rover moves upon making its first turn. As the rover travels, it sweeps out an angle $\gamma$ on this great circle. 

We adopt the usual convention of $(\phi, \theta)$ to be the pair of angles representing the CCW rotation from the $x = 0$ axis in the $x-y$ plane and the angle from the point to the the $x-y$ plane respectively. See the small diagram in the corner above.

Using a bit of spherical geometry, we can write the position of the point/rover as 

$$
\begin{align}
x & = \cos \gamma = \cos \theta \cos \phi \\ 
y & = \sin \gamma \cos \alpha = \cos \theta \sin \phi \\ 
z  & = \sin \gamma \sin \alpha = \sin \theta
\end{align}
$$

Dividing the second by the first, we find

$$
\tan \gamma \cos \alpha = \tan \phi \implies \boxed{\frac{\tan \phi}{ \tan \gamma} = \cos \alpha}
$$

Now, note that when the rover completes its second leg, we have $\gamma = s$ and $\phi = \frac{2 \pi - s}{2}$ since we still need to complete the last leg and by symmetry we need to complete half of the remaining loop around the $x-y$ great circle. Adding these two constraints and plugging in $\alpha = \pi/3$, we arrive the governing equation :

$$
 \boxed{\frac{\tan(\pi - s/ 2)}{ \tan s}  = 1/2}
$$

which has first solution of $s = 2 \arctan \sqrt{2}$. However recall we *normalized* so need to scale up by the radius of $R = 1000$ to arrive at $$ \boxed{ s = 2000 \arctan \sqrt{2}}$$




# Part 2

If we now want to limit $s$ to be no more than $100 R$, then in our normalized version we seek solutions in $s$ that have $s \leq 100$, plotting $\frac{\tan(\pi - s/ 2)}{ \tan s}$ over the range $s \in [0, 100]$ we see:





Which has pairs of solutions appearing in $2 \pi$ increments.

$$s_k = 2 \arctan \sqrt{2} + 2 \pi k \quad k \in \mathbb{Z}_{\geq 0}$$

$$s'_k = -2 \arctan \sqrt{2} + 2 \pi k \quad k \in \mathbb{Z}_{\geq 1}$$

Each of which permit $16$ solutions in $k$ before the length exceeds $100R$. Therefore, we suppose there are $\boxed{32}$ solutions in $s$ that allow the rover to return such that $s$ does not exceed $100 R$
