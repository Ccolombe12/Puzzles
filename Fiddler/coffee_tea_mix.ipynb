Lets focus on the amount of coffee in the left and right container after each iteration.

Let $\ell_i$ and $r_i$ be the volume of coffee in the left and right container respectively after the $i$-th operation. Let us assume we normalize the initial volume of liquid in each container to be unit valued. 

Observe that when we pour liquid from containers $A$ to $B$, the concentration of each beverage in $A$ remains the same and the concentrations in container $B$ are updated. Now suppose that on iteration $i$ we choose to perform the swapping operation with volume $v_i \in [0, 1]$. We can then use the above observation to write the updated volume of coffee in each as:

$$
\begin{equation}
\begin{aligned}
\ell_{i + 1} & = \ell_i(1 - v_i) + v_i \frac{r_i + v \ell_i}{1 + v_i} \\
r_{i + 1} & = \frac{r_i + v \ell_i}{1 + v_i}
\end{aligned}
\end{equation}
$$

where the initial base cases are $\ell_0 = 1$ and $r_0 = 0$.

We can simplify this by noting that $r_i = 1 - \ell_i$ (the total volume of coffee must be 1), then we obtain:

$$
\begin{equation}
\boxed{\ell_{i + 1}  = \ell_i(1 - v_i) + v_i \frac{1 - \ell_i + v_i \ell_i}{1 + v_i} }
\end{equation}
$$


Which tells us how the volume of coffee in the left container changes on each iteration. We would like to minimize this by choosing some sequence of pours $v_1, v_2 \ldots$. What should this sequence be?

We can start by looking at how the volume of coffee changes on each iteration. Dropping the index notation for a moment, the change in volume starting at $\ell$ is given by

$$

\begin{equation}
\boxed{\Delta \ell = - v \ell + v \frac{1 - \ell + v \ell}{1 + v}}
\end{equation}
$$

We can see that in order to decrease the volume of coffee in the left container ($\Delta\ell < 0$) we must have

$$
\begin{equation}
\begin{aligned}
\Delta\ell &=  -v \ell + v \frac{1 - \ell + v \ell}{1 + v}  < 0 \\
 \implies & \\
  &\boxed{\ell > 1/2}
\end{aligned}
\end{equation}
$$

That is, we can *decrease* the volume of coffee if and only if $\ell > 1/2$. In other words, we can only decrease the volume of coffee in the left container if there is more than half of the coffee currently in it. This implies that if $\ell < 1/2$ ever during the sequence of operations, we should end the process.

Can we we ever get $\ell  < 1/2$? It turns out the answer is no! We can prove that by showing that for fixed $\ell \geq 1/2$, we can never find a $v$ that causes the $\ell$ to dip below $1/2$.

Note that if $\ell_i = 1/2$, the equation for $\ell_{i+ 1}$ reduces to

$$
\begin{equation}
\begin{aligned}
\ell_{i + 1}  & = \ell_i(1 - v_i) + v_i \frac{1 - \ell_i + v_i \ell_i}{1 + v_i}\\
& = \frac{1}{2}(1 - v_i) + v_i \frac{\frac{1}{2} + v_i \frac{1}{2} }{1 + v_i}\\
& = \frac{1}{2}(1 - v_i) + \frac{1}{2}v_i \frac{1+ v_i }{1 + v_i}\\
& = \frac{1}{2}(1 - v_i) + \frac{1}{2}v_i\\
& = \frac{1}{2}
\end{aligned}
\end{equation}
$$

implying there it is now impossible to change the volume of coffee in the left container from its fixed point of $1/2$.

On the other hand, suppose $i + 1$ is the earliest instance where $\ell_{i + 1} < 1/2$. Then $\ell_i > 1/2$. For what $v \in [0, 1]$ can we obtain $\ell_{i + 1 } < 1/2$? Looking at our difference equation again, we need $v$ to satisfy

$$
\begin{equation*}
 -v \ell_i + v \frac{1 - \ell_i + v \ell_i}{1 + v} < 1/2 - \ell_i \qquad \text{(Decrease covers the current distance to $1/2$)}
\end{equation*}
$$

However, if we simplify this inequality, we find that this condition only holds when $v > 1$, which is not possible. Therefore there can not be an earliest moment in which the process dips below $1/2$ and thus $\ell_i \geq 1/2$ for all $i$.

Therefore, $\boxed{\ell* =1 /2}$ is the best we can hope to achieve. Furthermore, it is easy to verify that it can be achieved in one step using $v_1 = 1$.

How does the volume of coffee change each time? 

Now 

$$
\begin{equation}
\ell_{i+ 1} = \ell_i - v_i \left(\frac{\ell_i}{1 + v_i}\right)  = \ell_i \left(1 - \frac{v_i}{1 + v_i}\right)
\end{equation}
$$

Which implies that after $N$ operations with the chosen volumes $v_1, \ldots v_N$, and $\ell_0 = 1$ we get:

$$
\begin{equation}
\ell_{N} = \prod_{i = 1}^N\left(1 - \frac{v_i}{1 + v_i}\right)
\end{equation}
$$
with the constraint that $\sum_{i}v_i \leq 1$. By symmetry, we can argue that at optimality $v_1 = v_2 = 
\ldots =v_N = 1/N$, and we should then expect:

$$
\begin{equation}
\ell_{N} = \left(1 - \frac{1/N}{1 + 1/N}\right)^N = \left(1 - \frac{1}{N + 1}\right)^N 
\end{equation}
$$

Which is decreasing in $N$ and in the limit converges to $1/e$.
$$
\begin{equation}
\boxed{\ell^* = \lim_{N \to \infty} \left(1 - \frac{1}{N + 1}\right)^N  = \frac{1}{e}}
\end{equation}
$$
