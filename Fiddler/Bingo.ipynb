# Part 2: For an $N \times N$ BINGO game (where only tiles on your board are called) how long to BINGO on average?


```python
# assume nxn board.
from itertools import combinations
from fractions import Fraction


def check_bingo(board):
    # checks if the current board has a BINGO
    N = len(board)

    def row_check(r):
        # check if row r is a BINGO
        return all(board[r])

    def col_check(c):
        # check if column c in a BINGO
        return all(board[r][c] for r in range(N))

    def diag_checks():
        # check if either diagonal is a BINGO
        main_diag_chk = all(board[i][i] for i in range(N))
        anti_diag_chk = all(board[i][N - 1 - i] for i in range(N))
        return main_diag_chk or anti_diag_chk

    return (
        any(row_check(r) for r in range(N))
        or any(col_check(c) for c in range(N))
        or diag_checks()
    )


def make_board(N):
    # returns an empty BINGO board
    return [[0] * N for _ in range(N)]


def get_location(x, N):
    # get the location (r, c) of number x in the board.
    return x // N, x % N


def populate_board(N, locations):
    # place the chosen locations, and the free-space, on the board
    board = make_board(N)
    r_free, c_free = N // 2, N // 2
    board[r_free][c_free] = 1
    for x in locations:
        r, c = get_location(x, N)
        board[r][c] = 1
    return board


def get_possible_placements(N, k):
    # fetches a list of all possible ways to place k markers on the board
    possible_elements = [
        x for x in range(N**2) if x != N ** 2 // 2
    ]  # elements that are not the free space
    return combinations(possible_elements, k)


def prob_no_bingo(N, k):
    # computes the probablity that if we place k markers + free space on a board
    # we will NOT have a bingo.
    total_placements = 0
    non_bingo_placements = 0
    placements_iter = get_possible_placements(N, k)

    for locations in placements_iter:
        total_placements += 1
        cur_board = populate_board(N, locations)
        if not check_bingo(cur_board):
            non_bingo_placements += 1

    return (
        Fraction(non_bingo_placements, total_placements) if total_placements != 0 else 0
    )


def get_expected_bingo_time(N):
    # make use of the i# dentity for non-negative discrete RV E[X] = Sum(Pr[X > k] for k >=0)
    min_k = N - 1  # min number of markers needed
    max_k = (N) ** 2 - 1  # max number of markers needed
    total = min_k  # sum from 0 ... min_k - 1
    for k in range(min_k, max_k + 1):
        print(f"checking {k = }")
        delta_p = prob_no_bingo(N, k)
        if delta_p == 0:
            break
        total += delta_p
    return total
```

```python
E3 = get_expected_bingo_time(3)
print(f"It takes {E3} ~ {float(E3):.3f} markers on avg for N = 3")
```

```python
E5 = get_expected_bingo_time(5)
print(f"It takes {E5} ~ {float(E5):.3f} markers on avg for N = 5")
```
