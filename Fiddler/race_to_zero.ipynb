# Part 1

Let us start with some notation. Suppose the distance to be traveled during the race is $D$ and let the initial speed be $v_0$. Running at a $p$-minute pace would then imply that $v_0 = D/p$ etc. Furthermore, suppose the runner increases their speed by $\alpha > 0$ percent each time they speed up.

Now, we can break the race into segments (zero indexed) for which the runner's speed is constant. In the $k$-th segment, their speed is $v_k = (1 + \alpha)v_{k-1}  = (1 + \alpha)^k v_0$ and they cover distance $d_k = \frac{D}{2^{k + 1}}$. Thus the time spent in this segment is 

$$
\begin{align*}
t_k &= \frac{d_k}{v_k} = \frac{D / 2^{k + 1}}{(1 + \alpha)^k v_0}\\
& = \frac{D}{2v_0} \left(\frac{1}{2(1 + \alpha)}\right)^k.
\end{align*}
$$


Now we can add up the total time from all the segments to get the total race time

$$
\begin{align*}
T & = \sum_{k = 0}^\infty \frac{D}{2v_0} \left(\frac{1}{2(1 + \alpha)}\right)^k\\
& = \frac{D}{2v_0}  \sum_{k = 0}^\infty  \left(\frac{1}{2(1 + \alpha)}\right)^k\\ 
& = \frac{D}{2v_0} \frac{1}{1 - \frac{1}{2(1 + \alpha)}} \\
& = \frac{D}{v_0} \frac{1 + \alpha}{1 + 2 \alpha } \\
& = p\;  \frac{1 + \alpha}{1 + 2 \alpha }
\end{align*}
$$

Now if $\alpha = 1/10$ and $p = 24$ (the slower initial pace), then we find $\boxed{T = 22}$ which implies that this negative split strategy would lead to a faster time than the flat 23 minute pace! (provided you can surpass light-speed).

____

Now we want to define a continuous speed function $v(x)$ that represents the speed of the runner when they have completed $x$ fraction of the race. It will align with the speed of the previous case of $v_k  = v_0(1 + \alpha)^k$ when $x = 1 - \frac{1}{2^k}$ for non-negative integers $k$. We can use this to determine a function that satisfies the requirements of $v(x)$


If we just consider the exponent $k$ on the $(1 + \alpha)^k$ term we just mentioned, we develop a relationship between $x$ and $k$.

For examples, 
$$
\begin{align}
x = 0  & \implies k = 0 \\
x = 1/2 &  \implies k =1 \\
x = 3/4 &  \implies k =2
\end{align} 
$$
etc.

From this, we can reason the relation 

$$ 1 - \frac{1}{2^k} = x$$

which implies 

$$ k = - \log_2(1 - x)$$

and a functional form of 
$$
\begin{equation}
\boxed{v(x) = v_0 (1 + \alpha)^{- \log_2(1 - x)}}
\end{equation}
$$

Now, to determine the time it takes to finish the race, we must consider the time it takes to cross the infinitesimally small distance $dx$ at position $x$, which we denote $dt$.
We can reason 
$$
\begin{align*}
dt & = \frac{dx}{v(x)} \\ 
& = \frac{dx}{v_0 (1 + \alpha)^{- \log_2(1 - x)}}
\end{align*}$$

Adding up all of these time intervals, (and letting $D = 1$ WLOG), we find

$$
\begin{align*}
T & = \int_{0}^1 \frac{dx}{v_0 (1 + \alpha)^{- \log_2(1 - x)}}\\ 
& = p \int_{0}^1 \frac{dx}{(1 + \alpha)^{- \log_2(1 - x)}}\\ 
& = p \frac{\log (2)}{\log (2 ( 1 + \alpha))}
\end{align*}$$



Which implies that if run at the inital pace of $p = 24$ then it will take $\boxed{T = 24 \frac{\log (2)}{\log \left(\frac{11}{5}\right)} \approx 21.0988}$
