We can use dynamic programming to compute the number of possible combinations.
The idea is to iterate through our list of possible numbers, check if they match the prefix of the remaining key,
if they do, then we can get the new remaining string and continue


```python
from functools import cache


@cache
def count(cur_str: str, characters: tuple[str]):
    """Returns the number of ways to write cur_str using characters
    Args:
        cur_str: The remaining string
        characters: Tuple of characters we can use to make the string
    """
    if cur_str == "":
        return 1
    cnt = 0
    for c in characters:
        if cur_str.startswith(c):
            cnt += count(cur_str[len(c) :], characters)
    return cnt


# Part 1
characters = ("I", "II", "III")
s = "I" * 10
ans1 = count(s, characters)
print(f"The number of ways to write {s} using {characters} is {ans1}")
```

```python
# now do the same with a much more expanded set of characters.
characters = ("I", "II", "III", "IV", "V", "VI", "VII", "VIII")
s = "IIIVIIIVIIIVIII"
ans2 = count(s, characters)
print(f"The number of ways to write {s} using {characters} is {ans2}")
```
