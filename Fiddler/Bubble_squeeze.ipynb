# Part 2

Now we would like to find the limiting area when $N$ is very large.


One observation to make early on is that

- if we were to place the center of the initial circle on a vertex of a unit-hexagonal lattice, then all subsequent points would lie on the lattice points.
- Moreover, For a given number of circles placed $N$, the lattice points representing the centers of these circles will form concentric hexagons about the central lattice point.
- Which suggests for larger and larger $N$ we might expect to see a _hexagonal_ shape emerge.


Let's suppose now that $N$ is a number that can be represented as the sum of the number of vertices on concentric hexagons with side lengths increasing by 1. Thus we will have some arrangement that looks like the one above. In this case, we can upper bound the total area by the "bounding hexagon" (shown in black). Given that there are $N$ circles, let $s(N)$ be the side-length of the hexagon that bounds it. If $N$ is the sum of the vertices on these hexagonal rings with side-length up to $s$, then we can write $s$ in terms of $N$ using the following logic

$$
\begin{align*}
N &= 1 + \sum_{k = 1}^s 6 k \\
N & = 1 + 6 \left(\frac{1}{2} s( s + 1)\right)\\
\implies & \\
s(N) & = \frac{1}{6} \left(\sqrt{12 N - 3}-3\right)\\
s(N) & \approx \frac{\sqrt{3 N}}{3} && (N \gg 1).
\end{align*}
$$

In the last step, we use the approximation that $N$ is very large.


The "bounding" hexagon for the circles that lie on the hexagonal lattice points at at "distance" $s$ from the center, has side-length $s + 1$ and thus we can get an approximation (and upper bound) for the area of the union of the circles $A(N)$ as

$$
\begin{align*}
A(N) & \leq \frac{3 \sqrt{3}}{2} \left(s(N) + 1\right)^2 \\
& \approx \frac{3 \sqrt{3}}{2} \left(\frac{\sqrt{3 N}}{3}\right)^2  && (N \gg 1)\\
& = \frac{\sqrt{3}}{2} N
\end{align*}
$$


This is not entirely rigourous as we may have violated the inequality when applying our approximation but the quantities will only be off by a linear factor of $s$ and in the limit can be ignored. Furthermore, this argument was predicated on the notion that $N$ was one of these "sum of hexagon vertices" numbers. For other $N$, there is still the same "bounding" hexagon, the last layer of which is just not quite as filled out. The "inner" hexagon (hexagon with side length $\lfloor s(N) \rfloor$) only differs from the "bounding" hexagon by a linear factor of $s$ and thus the two approach the same limit for large $N$. We can use a squeeze theorem type argument to then argue that the limiting area $\frac{\sqrt{3}}{2} N$ is reasonable for sufficiently large $N$.
