# Part 1

We are tasked with determining the length $L$ for which the tower below will no longer be able to stand on its base.

The tower will tip once its horizontal center-of-mass no longer falls to the right of the leftmost edge of the support. 
We can handle this by considering the block in two pieces. The right and static piece of size $1 \times 2$ and the left and growing piece of size $1 \times L$. We will assume a uniform and constant density per unit volume added of $\rho = 1$ units of mass/unit vol.

Let use define $x = 0$ as the location on the $x$-axis where the left edge of the base is. We are interested when the $x_{COM}$ of the the whole piece crosses this line. 

 The $x_{COM}$ of the right piece is at $1/2$ with weight $2$, the left piece, for fixed $L$ has $x_{COM}$ at $-L/2$ with mass $L \cdot \rho =  L$. Setting the entire COM to 0, we have 


$$
\begin{equation}
 - \frac{L^2}{2} + 1 = 0 
\end{equation}
 $$ 

which of course then implies that $\boxed{L = \sqrt{2}}$ is the tipping point for our block. 

# Part 2



Now we have a similar problem but with an annulus. We will generalize the problem a bit here and let's instead suppose the outer and inner radii are $R$ and $r$ respectively.
We will also slightly change up the $x$-axis convention. Let $x = 0$ now be the center of the annulus. Here, the tipping point will now be at $x = r$ instead of $x = 0$.

Similar to before we need to find the point at which the $x_{COM}$ reaches $x = r$. For a contigous object, we can get the $x_{COM}$ using the following formula.

$$
\begin{equation}
x_{COM} = \frac{\int x \cdot dm }{M}
\end{equation}
$$

where $M$ is the total mass (which we ignored in the previous problem since we set this equation equal to zero) and we are summing the infinitesimal contributions from the mass in the object. We need to first find the mass of the object as a function of $\theta$. If we assume uniform and unit density as we did prior, this is equivalent to finding the area of the region, which we can do with some simple geometry to find 

$$
M = \frac{\theta}{2 \pi }\left(\pi R^2 - \pi r^2\right) = \theta\frac{R^2 - r^2}{2}
$$


Now we need to take the integral, if we which to polar coordinates $(r', \phi)$, this is a much easier task. Be careful to remember that when doing to polar, the $dA \to r' dr' d\phi$.
for a given region of area $dA$ the mass is $dm =\rho dA = dA = r' dr' d\phi$. The $x$ coordinate for a given position in polar coordinates is given by $r' \cos \phi$, and therefore our $x_{COM}$ integral becomes 

$$
\begin{equation}
\begin{aligned}
\int_{0}^\theta\int_{r}^R r' \cos \phi r' dr' d \phi & = \int_0^\theta \int_{r}^R (r')^2 \cos \phi dr' d \phi \\ 
& = \int_{0}^\theta \frac{R^3 - r^3}{3} \cos \phi d \phi\\
& = \sin \theta \frac{R^3 - r^3}{3}
\end{aligned}
\end{equation}
$$

Combining this with our earlier result, the $x_{COM}$ is 

$$
\begin{equation}
x_{COM} = \frac{2 \sin \theta}{3 \theta} \frac{R^3 - r^3}{ R^2 - r^2}
\end{equation}
$$

We need to choose $\theta$ so that that this becomes $r$. Which means we are after the solution in $\theta$ to the equation.

$$
\begin{equation}
\boxed{r = \frac{2 \sin \theta}{3\ \theta} \frac{R^3 - r^3}{ R^2 - r^2}}
\end{equation}
$$

Which if we solve for $\theta$ for our given $R = 2$ and $r = 1$, we find $\boxed{\theta \approx 1.555370481796972}$, which is just ever so shy of $\pi/2$. For fun, if we plot this critical $\theta$ as a function of $r/R \in (0, 1)$, we recover the plot



Which tells us the following:
* For $r \approx R$, the base of the shape is so relatively small it almost immediately tips over.
* For $r \ll R$, we essentially have nearly a semicircle so it is very stable and can nearly touch the ground on the other side before tipping.
